```python
import pandas as pd
import numpy as np
%matplotlib inline
pd.options.display.float_format = '${:,.2f}'.format
```

```python
props = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

```python
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

```python
merged = pd.merge(props, contribs, on="calaccess_committee_id")
```

```python
def mktable(table, groupby):
    return table.groupby(groupby).amount.sum().reset_index().sort_values("amount", ascending=False)
    
mktable(merged, "contributor_lastname").head(20)
```

```python
merged.groupby("contributor_lastname").amount.sum().reset_index().sort_values("amount", ascending=False).head(20)
```

```python
merged.groupby(["contributor_lastname", "committee_name_x", "prop_name"]).amount.sum().reset_index().sort_values("amount", ascending=False).head(16)
```

```python
merged[merged.contributor_lastname == "AIDS HEALTHCARE FOUNDATION"]
```

```python
aids_foundation = merged[merged.contributor_lastname == "AIDS HEALTHCARE FOUNDATION"]
aids_foundation.groupby(["prop_name", "committee_name_x", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
aids_foundation.groupby("date_received").amount.sum().reset_index().sort_values("date_received", ascending=True)
```

```python
dates = aids_foundation.groupby("date_received").amount.sum().reset_index().sort_values("date_received", ascending=True)
timeline = dates.amount.plot.line().set_xticklabels(aids_foundation.date_received)
```

```python
dates.to_csv("dates.csv")
```
